```python
import matplotlib.pyplot as plt
import numpy as np
conocimientos = []
conocimientosSinRepetir = []
lenguajesP = np.array(['C', 'C++', 'Java', 'Python', 'PHP', 'Go', 'C#', 'JavaScript', 'R', 'SWIFT'])
cantidad = []
file = open('cargos.csv','r', encoding='UTF-8')
file.readline()
for linea in file:
    linea = linea.strip()
    nombre_cargo, ciudad, años_experiencia, requisitos = linea.split('|')
    l_requisitos = requisitos.split(',')
    for elemento in l_requisitos:
        conocimientos.append(elemento.strip())
file.close()
for x in conocimientos:
    if x not in conocimientosSinRepetir:
        conocimientosSinRepetir.append(x)
        cantidad.append(1)
    else:
        r = conocimientosSinRepetir.index(x)
        cantidad[r] += 1

lenguajesProgramacion = []
cantidadLP = []

for i in range(len(conocimientosSinRepetir)):
    if conocimientosSinRepetir[i] in lenguajesP:
        lenguajesProgramacion.append(conocimientosSinRepetir[i])
        cantidadLP.append(cantidad[i])

fig = plt.figure(figsize =(8.5, 8.5))
plt.pie(cantidadLP, labels=tuple(lenguajesProgramacion), autopct="%0.1f %%", 
colors = ( "#009933", "#FF9933", "#339933", "#006666", "#FF9999","#FFFF99", "#66FFFF","#CC9900","#CCCC00","#A04000"))
plt.show()
```
