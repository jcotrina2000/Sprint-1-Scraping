```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
cargos = pd.read_csv('jobs_expertini.csv', sep=';')
cargos 
```

```python
# conteo de numero de empleos por ciudad 
temp = cargos[['empresa','titulo']]
countt= temp.groupby('empresa').count()
print(countt)
countt.plot(kind='barh', figsize=[20,8])
plt.xlabel('numero de empleos')
plt.title('Cantidad de empleos posteados por cada empresa')
countt.plot.pie(subplots=True,figsize=(16, 18), autopct="%0.1f %%")
```

```python
import plotnine as p9
myplot = (p9.ggplot(data=cargos, mapping=p9.aes(x='provincia', y='empresa')) +
              p9.geom_point())

# convierte el output de plotnine a un objeto de matplotlib
my_plt_version = myplot.draw()

# Realiza más ajustes al gráfico con matplotlib:
p9_ax = my_plt_version.axes[0] # cada subgráfico es un ítem en una lista
p9_ax.set_xlabel("Hindfoot length")
p9_ax.tick_params(labelsize=10, pad=16)
p9_ax.set_title('Empresas disponibles en cada provincia', fontsize=15)
plt.show() # esto no es necesario en Jupyter Notebooks
```
